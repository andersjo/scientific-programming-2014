`Networkx`
====


```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import codecs
import numpy as np
import operator
import pandas as pd
```

NetworkX let's us define graphs very easily, and then gives us a lot of different algorithms to analyze it. The steps are simple:

1. select a graph type: undirected (`Graph`) or directed (`DiGraph`) graph

2. populate the graph with nodes and/or edges. If we add an edge, the nodes are automatically added
    
    * we can add individual nodes (`add_node(u)`) or edges (`add_edge(u, v)`), or 
    * add them in batches (`add_nodes_from(x)`, `add_edges_from(x)`)

Examples
---

First, we read in a file that defines the graph. Here, we use the reviews each user has left for a company in our data. Each review is one edge between a user and a company. The rating is the strength of the edge. Since this is a network, and since companies don't review users, we opt for a directed graph here.

```python
G = nx.DiGraph()

for i, line in enumerate(codecs.open('trustpilot_graph.txt', encoding='utf-8')):
    line = line.strip()
    uid, cid, rating = line.split('\t')

    # add each endpoint and set a color and type attribute (names arbitrary)
    G.add_node(uid, color='green', user_type='user')
    G.add_node(cid, color='red', user_type='company')
    # add a link
    G.add_edge(uid, cid, weight=float(rating))

    if i == 1000:
        break

G.size()
```

We can now enrich our graph by adding information about each node. We use some standard measures that assess how important a node is

* degree centrality: this measures to how many other nodes a node $u$ is connected

* betweenness centrality: this measures how often the node is part of a shortest path between other nodes

* Page Rank: this is the algortihm used by Google (and developed by one of its fonders, Larry Page). It takes into account how many nodes link to $u$ and how important those other nodes are.



```python
page_rank = nx.pagerank(G)
nx.set_node_attributes(G, 'page_rank', page_rank)
sorted_pr = sorted(page_rank.items(), key=operator.itemgetter(1), reverse=True)

degree_centrality = nx.degree_centrality(G)
nx.set_node_attributes(G, 'degree_centrality', degree_centrality)
sorted_degree_centrality = sorted(degree_centrality.items(), key=operator.itemgetter(1), reverse=True)

betweenness = nx.betweenness_centrality(G)
nx.set_node_attributes(G, 'betweenness', betweenness)
sorted_degree_centrality = sorted(betweenness.items(), key=operator.itemgetter(1), reverse=True)

for node_id, pr_degree in sorted_pr[:10]:
    print node_id, G.node[node_id]
```

Note that for a directed graph, betweenness is 0 for all nodes, since there are no path longer than 1. Similarly, users have no Page rank, since they don't have any incoming nodes pointing to them.


We can now plot theses measures against each other in order to get a visual of which nodes are more important according to which measures. We label only the top 10.

```python
bw = [G.node[nid]['betweenness'] for nid in G.nodes_iter()]
pr = [G.node[nid]['page_rank'] for nid in G.nodes_iter()]
dc = [G.node[nid]['degree_centrality'] for nid in G.nodes_iter()]
user_type = [G.node[nid]['user_type'] for nid in G.nodes_iter()]

tp_data = pd.DataFrame({'id':G.nodes(), 'betweenness':bw, 'page_rank':pr, 'degree_centrality':dc, 'user_type':user_type})

def scatter_plot_importance(g, m1, m2, top=10):
    fig, ax = plt.subplots()
    fig.set_size_inches(20,8)

    g.plot(x=m1, y=m2, style='o', ax=ax)

    aggregate = zip(g[m1] + g[m2], g['id'], g[m1], g[m2])
    aggregate.sort(reverse=True)
    for _, label, x, y in aggregate[:top]:
        plt.annotate(label,
                        xy = (x, y), xytext = (-20, 20),
                        textcoords = 'offset points', ha = 'right', va = 'bottom',
                        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))


scatter_plot_importance(tp_data, 'betweenness', 'degree_centrality')
scatter_plot_importance(tp_data, 'page_rank', 'degree_centrality')
```

We can also make separate plots for users and companies.

```python
scatter_plot_importance(tp_data[tp_data['user_type'] == 'user'], 'page_rank', 'degree_centrality')
scatter_plot_importance(tp_data[tp_data['user_type'] == 'company'], 'page_rank', 'degree_centrality')
```

Let's look at the top 3 users and their connections.

```python
super_users = tp_data[tp_data['user_type'] == 'user'].sort('degree_centrality').tail(3).id.tolist()
subset = super_users
for x in super_users:
    subset.extend(G[x].keys())

# print subset

S = nx.subgraph(G, subset)
fig, ax = plt.subplots()
fig.set_size_inches(20,20)

nx.draw(S, node_color=[S.node[x]['color'] for x in S.nodes_iter()], ax=ax)
plt.show()
```
